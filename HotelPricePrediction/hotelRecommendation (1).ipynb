```python
df['more'] = df['more'].str.split(',')
```

```python
def extract_adults(details):
    for item in details:
        if 'adult'  in item:
            return int(item.split()[0])
    return None

# Apply the function to create a new column 'Adults'
df['Adults'] = df['more'].apply(extract_adults)
```

```python
def extract_adults(details):
    for item in details:
        if 'child'  in item:
            return int(item.split()[0])
    return 0

# Apply the function to create a new column 'Adults'
df['child'] = df['more'].apply(extract_adults)
```
